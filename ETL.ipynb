```python
import pandas as pd
import requests
```

```python
url = "https://arbeitnow.com/api/job-board-api"

payload = {}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

data = response.json()
```

```python
df = pd.DataFrame(data['data'])
```

```python
df['tags'] = df['tags'].apply(lambda x:",".join(x))
```
