```python
import pandas as pd
import requests
```

```python
url = "https://coronavirus.m.pipedream.net/"

payload = {}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)

data = response.json()
```

```python
df = pd.DataFrame(data['rawData'])
```

```python
df
```

```python
df_turkiye = df[df["Combined_Key"] == "Turkey"]
```
